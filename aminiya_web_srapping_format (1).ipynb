```python
import requests
from bs4 import BeautifulSoup
import csv


unique_links = set()


for page_number in range(501, 1001):
    url = f"https://aminiya.ng/sabbin-labarai/page/{page_number}"
    response = requests.get(url)
    soup = BeautifulSoup(response.content, "html.parser")
    ul_element = soup.find('ul', class_='hero_latest_list')
    a_elements = ul_element.find_all('a')

    for a_element in a_elements:
        link = "https://aminiya.ng/" + a_element['href']
        unique_links.add(link)

with open('scraped_content1000.csv', 'w', newline='', encoding='utf-8') as csvfile:
    csv_writer = csv.writer(csvfile)
    csv_writer.writerow(['Link', 'Content'])


    with open('scraped_content1000.txt', 'w', encoding='utf-8') as textfile:
        for link in unique_links:
            link_response = requests.get(link)
            link_soup = BeautifulSoup(link_response.content, "html.parser")
            content_element = link_soup.find('div', class_='post-middle article-container post')


            if content_element:
                content = content_element.get_text()
                content = content.strip()
                csv_writer.writerow([link, content])
                textfile.write(f"{content}")
                print(f"Completed link: {link}")

print("Scraped data saved to scraped_content.txt and scraped_content.csv")
```
