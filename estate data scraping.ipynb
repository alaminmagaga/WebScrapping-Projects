```python
import requests
from bs4 import BeautifulSoup
from selenium import webdriver

url = "https://housing.com/in/buy/searches/P38f9yfbk7p3m2h1f"
response = requests.get(url)
soup = BeautifulSoup(response.content, "html.parser")

data = []

for item in soup.find_all("div", {"class": "css-zrd0bv"}):
    property_price_range = item.find("div", {"class": "css-18rodr0"}).text.strip()
    price = item.find("span", {"class": "css-nin1gj"}).text.replace("\n", "").strip()
    Developer = item.find("h2", {"class": "css-zxue4u"}).text.replace("\n", "").strip()
    Location = item.find("div", {"class": "css-4in8py"}).text
    area = item.find_all("div", {"class": "css-ebj250"})[1].text
    link = "https://housing.com/" + item.find_all("a", {"class": "css-0"})[0]['href'] 
    

    link_response = requests.get(link)
    link_soup = BeautifulSoup(link_response.content, "html.parser")
    
    # Find the relevant content on the link page
    description = link_soup.find_all("div", {"class": "css-13t1l0h"})[0].text.strip()
    
    print(f"Property Price Range: {property_price_range}")
    print(f"Price: {price}")
    print(f"Developer: {Developer}")
    print(f"Location: {Location}")
    print(f"Area: {area}")
    print(f"Link: {link}")
    print("")
    
    # Use Selenium to retrieve the contact details of each housing listing from its details page
    driver = webdriver.Chrome()
    driver.get(link)

    anchor_soup = BeautifulSoup(driver.page_source, "html.parser")
    contact_details = anchor_soup.find("div", {"class": "seller-phone _1q731trj _j3mgvx _h31y44 _csbfng _c81fwx"})
    if contact_details:
        contact_details_text = contact_details.text.strip()
    else:
        contact_details_text = "Not available"

    driver.quit()
    
    data.append([property_price_range, price, Developer,  Location, area, link, contact_details_text])

# Create a dataframe from the data list
df = pd.DataFrame(data, columns=["Property Price Range", "Price", "Developer", "Location", "Area","Link", "Contact Details"])

# Save the dataframe as a CSV file
df.to_csv("housing_data.csv", index=False)
```
