```python
import requests
from bs4 import BeautifulSoup
import csv

base_url = "https://www.arewaradio.com"
archive_url = "/news/labarai-news/archive/"

response = requests.get(base_url + archive_url)
soup = BeautifulSoup(response.content, "html.parser")

ul_element = soup.find('ul', class_='gm-sec divide-rows')

link_elements = ul_element.find_all('a')

# Create a CSV file and write header
csv_file = open('scraped_news.csv', 'w', newline='', encoding='utf-8')
csv_writer = csv.writer(csv_file)
csv_writer.writerow(['Link', 'Content'])

count = 0  # To keep track of the number of scraped news

for link in link_elements:
    if count >= 10:
        break

    href = link.get('href')
    if href.startswith("/"):
        full_link = base_url + href
        print(f"Link: {full_link}")

        sub_response = requests.get(full_link)
        sub_soup = BeautifulSoup(sub_response.content, "html.parser")

        sub_ul_element = sub_soup.find('ul', class_='gm-sec divide-rows')
        sub_link_elements = sub_ul_element.find_all('a')

        for sub_link in sub_link_elements:
            if count >= 10:
                break

            sub_href = sub_link.get('href')
            if sub_href.startswith("/"):
                full_sub_link = base_url + sub_href
                print(f"Sublink: {full_sub_link}")

                sub_sub_response = requests.get(full_sub_link)
                sub_sub_soup = BeautifulSoup(sub_sub_response.content, "html.parser")

                # Find and print the content of the sublink, excluding aiir-c-news-extra
                content_element = sub_sub_soup.find('article', class_='gm-group')
                if content_element:
                    content = ""
                    for paragraph in content_element.find_all('p'):
                        if 'aiir-c-news-extra' in paragraph.get('class', []):
                            continue
                        content += paragraph.get_text() + '\n'

                    print(f"Content: {content.strip()}")

                    # Write data to CSV file
                    csv_writer.writerow([full_sub_link, content.strip()])

                    count += 1  # Increment the count

# Close CSV file
csv_file.close()
```
