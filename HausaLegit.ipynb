```python
import requests
from bs4 import BeautifulSoup
import csv
import logging
from google.colab import drive


drive.mount('/content/drive')


logging.basicConfig(filename='/content/drive/My Drive/scraping_log.txt', level=logging.ERROR, format='%(asctime)s - %(levelname)s: %(message)s')

unique_links = set()

for page_number in range(4001,4570):
    url = f"https://hausa.legit.ng/latest/?page={page_number}"
    try:
        response = requests.get(url)
        response.raise_for_status()
    except requests.exceptions.RequestException as e:
        logging.error(f"Error while requesting page {url}: {e}")
        continue

    soup = BeautifulSoup(response.content, "html.parser")
    ul_element = soup.find("div", class_="js-articles")

    if ul_element is None:
        logging.error(f"No 'js-articles' div found on page {url}")
        continue

    a_elements = ul_element.find_all('a')

    for a_element in a_elements:
        link = a_element['href']
        unique_links.add(link)

csv_path = '/content/drive/My Drive/HausaNLPDataset/content4570.csv'
txt_path = '/content/drive/My Drive/HausaNLPDataset/content4570.txt'

with open(csv_path, 'w', newline='', encoding='utf-8') as csvfile:
    csv_writer = csv.writer(csvfile)
    csv_writer.writerow(['Link', 'Content'])

    with open(txt_path, 'w', encoding='utf-8') as textfile:
        for link in unique_links:
            if "legit.ng" not in link:
                continue

            if "/news/" in link:
                continue

            try:
                link_response = requests.get(link)
                link_response.raise_for_status()
            except requests.exceptions.RequestException as e:
                logging.error(f"Error while requesting link {link}: {e}")
                continue

            link_soup = BeautifulSoup(link_response.content, "html.parser")
            content_element = link_soup.find('article', class_='post__main js-article-body')

            if content_element:
                content = content_element.get_text()
                content = content.strip()
                csv_writer.writerow([link, content])
                textfile.write(f"{content}\n")
                print(f"Completed link: {link}")

print("Scraped data saved to Google Drive: scraped_content.txt and scraped_content.csv")
```

```python
import requests
from bs4 import BeautifulSoup
import csv

unique_links = set()

for page_number in range(4001, 5001):
    url = f"https://hausa.legit.ng/latest/?page={page_number}"
    response = requests.get(url)
    soup = BeautifulSoup(response.content, "html.parser")
    ul_element = soup.find("div", class_="js-articles")
    a_elements = ul_element.find_all('a')

    for a_element in a_elements:
        link = a_element['href']
        unique_links.add(link)

with open('content5000.csv', 'w', newline='', encoding='utf-8') as csvfile:
    csv_writer = csv.writer(csvfile)
    csv_writer.writerow(['Link', 'Content'])

    with open('content5000.txt', 'w', encoding='utf-8') as textfile:
        for link in unique_links:
            if "legit.ng" not in link:
                continue

            if "/news/" in link:
                continue

            link_response = requests.get(link)
            link_soup = BeautifulSoup(link_response.content, "html.parser")
            content_element = link_soup.find('article', class_='post__main js-article-body')

            if content_element:
                content = content_element.get_text()
                content = content.strip()
                csv_writer.writerow([link, content])
                textfile.write(f"{content}\n")
                print(f"Completed link: {link}")

print("Scraped data saved to scraped_content.txt and scraped_content.csv")
```
