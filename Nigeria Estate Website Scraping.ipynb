```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = "https://nigeriapropertycentre.com/for-sale"
response = requests.get(url)
soup = BeautifulSoup(response.content, "html.parser") 
```

```python
all = soup.find_all("div", {"class": "wp-block property list"})
```

```python
data = []

for item in all:
    house_title=item.find("div", {"class": "wp-block-title hidden-xs"}).text.strip()
    house_subtitle=item.find("div", {"class": "wp-block-content clearfix text-xs-left text-sm-left"}).find("a").text.strip()
    address=item.find("address", {"class": "voffset-bottom-10"}).text.strip()
    amount=item.find("span", {"class": "pull-sm-left"}).text.strip()
    
    try:
        link = "https:" + item.find("div", {"class": "wp-block-content clearfix text-xs-left text-sm-left"}).find("a")["href"]
        
    except:
        print("Link not found")
        
  
    try:
        contractor_span = item.find("span", {"class": "marketed-by pull-right hidden-xs hidden-sm text-right"})
        contractor = contractor_span.text.strip().split("\n")[0]
        phone_number = contractor_span.find("strong").text.strip().replace("\xa0", "")
    except:
        contractor = None
        phone_number = None
    
    aux_info = item.find("div", {"class": "wp-block-footer"}).find("ul", {"class": "aux-info"})
    
    try:
        bedrooms = aux_info.find_all("li")[0].find_all("span")[0].text.strip()
    except:
        bedrooms = None
    
    try:
        bathrooms = aux_info.find_all("li")[1].find_all("span")[0].text.strip()
    except:
        bathrooms = None
    
    try:
        toilets = aux_info.find_all("li")[2].find_all("span")[0].text.strip()
    except:
        toilets = None
    
    try:
        parking_space = aux_info.find_all("li")[3].find_all("span")[0].text.strip()
    except:
        parking_space = None
    
    data.append({
        "House Title": house_title,
        "House Subtitle": house_subtitle,
        "Contractor": contractor,
        "Address": address,
        "Phone Number": phone_number,
        "Amount": amount,
        "Bedrooms": bedrooms,
        "Bathrooms": bathrooms,
        "Toilets": toilets,
        "Parking Space": parking_space,
        "Link": link
    })
    
    
    print(f"House Title: {house_title}")
    print(f"House Subtitle: {house_subtitle}")
    print(f"Contractor: {contractor}")
    print(f"Address: {address}")
    print(f"Phone Number: {phone_number}")
    print(f"Amount: {amount}")
    print(f"Bedrooms: {bedrooms}")
    print(f"Bathrooms: {bathrooms}")
    print(f"Toilets: {toilets}")
    print(f"Parking Space: {parking_space}")
    print(f"Link: {link}")
    
    print("")
    print("="*50)
```

```python
# save the scrape data in csv format
df = pd.DataFrame(data)
df.to_csv("nigeriapropertycentre.csv", index=False)
```

# scrape that using pagination by including the number of page

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

base_url = "https://nigeriapropertycentre.com/for-sale?page="

data = []

# number of page 1-3
for page in range(1, 3):
    url = base_url + str(page)
    response = requests.get(url)
    soup = BeautifulSoup(response.content, "html.parser")

    all = soup.find_all("div", {"class": "wp-block property list"})

    for item in all:
        house_title = item.find("div", {"class": "wp-block-title hidden-xs"}).text.strip()
        house_subtitle = item.find("div", {"class": "wp-block-content clearfix text-xs-left text-sm-left"}).find("a").text.strip()
        address = item.find("address", {"class": "voffset-bottom-10"}).text.strip()
        amount = item.find("span", {"class": "pull-sm-left"}).text.strip()

        try:
            link = "https:" + item.find("div", {"class": "wp-block-content clearfix text-xs-left text-sm-left"}).find("a")["href"]
        except:
            print("Link not found")

       
        try:
            contractor_span = item.find("span", {"class": "marketed-by pull-right hidden-xs hidden-sm text-right"})
            contractor = contractor_span.text.strip().split("\n")[0]
            phone_number = contractor_span.find("strong").text.strip().replace("\xa0", "")
        except:
            contractor = None
            phone_number = None

        aux_info = item.find("div", {"class": "wp-block-footer"}).find("ul", {"class": "aux-info"})

        try:
            bedrooms = aux_info.find_all("li")[0].find_all("span")[0].text.strip()
        except:
            bedrooms = None

        try:
            bathrooms = aux_info.find_all("li")[1].find_all("span")[0].text.strip()
        except:
            bathrooms = None

        try:
            toilets = aux_info.find_all("li")[2].find_all("span")[0].text.strip()
        except:
            toilets = None

        try:
            parking_space = aux_info.find_all("li")[3].find_all("span")[0].text.strip()
        except:
            parking_space = None

        data.append({
            "House Title": house_title,
            "House Subtitle": house_subtitle,
            "Contractor": contractor,
            "Address": address,
            "Phone Number": phone_number,
            "Amount": amount,
            "Bedrooms": bedrooms,
            "Bathrooms": bathrooms,
            "Toilets": toilets,
            "Parking Space": parking_space,
            "Link": link
        })
        
        print(f"House Title: {house_title}")
        print(f"House Subtitle: {house_subtitle}")
        print(f"Contractor: {contractor}")
        print(f"Address: {address}")
        print(f"Phone Number: {phone_number}")
        print(f"Amount: {amount}")
        print(f"Bedrooms: {bedrooms}")
        print(f"Bathrooms: {bathrooms}")
        print(f"Toilets: {toilets}")
        print(f"Parking Space: {parking_space}")
        print(f"Link: {link}")
    
        print("")
        print("="*50)
       
```

```python
df = pd.DataFrame(data)
df.to_csv("nigeriapropertycentre1.csv", index=False)
```
